```python
#過学習を抑える正則化
```

```python
import pandas as pd #pandas を pdとして扱う
import numpy as np #numpy を npとして扱う 
import matplotlib.pyplot as plt #matplotlib.pyplot を pltとして扱う
from sklearn.model_selection import train_test_split #sklearn.model_selection から train_test_splitをインポート
```

```python
dat = pd.read_csv("extended_boston.csv",index_col=0) #extended_boston.csvの読み込み、0行目を項目とする
X = dat.drop("y",axis=1) #ｙを落としてXに収納
y = dat.y #yのみ収納
```

```python
X #説明変数Xの表示
```

```python
y #目的変数ｙの表示
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0) #X,yのデータを毎回同じになるように訓練データとテストデータに分ける
```

```python
from sklearn.linear_model import LinearRegression #sklearn.linear_model から LinearRegressionをインポート
lin_reg = LinearRegression().fit(X_train,y_train) #線形回帰分析をX_trainを説明変数,y_trainを目的変数として実行
print(lin_reg.score(X_train,y_train)) #訓練データの精度を表示
print(lin_reg.score(X_test,y_test)) #テストデータの精度を表示
#訓練データの精度は高いが、テストデータの精度が低いので、過学習している！
```

```python
from sklearn.linear_model import Ridge #sklearn.linear_model から Ridgeをインポート
rid_reg = Ridge().fit(X_train,y_train) #リッジ回帰をX_trainを説明変数,y_trainを目的変数として実行
print(rid_reg.score(X_train,y_train)) #訓練データの精度を表示
print(rid_reg.score(X_test,y_test)) #テストデータの精度を表示
#訓練データの精度は低くなったが、テストデータの精度が上がったので、過学習を抑えより汎用的な予想ができるようになった。
```

```python
from sklearn.linear_model import Lasso #sklearn.linear_model から Lassoをインポート
las_reg = Lasso().fit(X_train,y_train) #ラッソ回帰をX_trainを説明変数,y_trainを目的変数として実行
print(las_reg.score(X_train,y_train)) #訓練データの精度を表示
print(las_reg.score(X_test,y_test)) #テストデータの精度を表示
#学習不足の典型的な例、どっちの精度も低い
```

```python
pd.DataFrame({"lin_reg":lin_reg.coef_,"rid_reg":rid_reg.coef_,"las_red":las_reg.coef_})
#線形、リッジ、ラッソ回帰の係数を表示
```

```python
from sklearn.linear_model import Lasso #sklearn.linear_model から Lassoをインポート
las_reg_001 = Lasso(alpha=0.01,max_iter=10000).fit(X_train,y_train) #正則加工の重みを0.01にして（少し弱め＝係数を少し多きくする）、繰り返し回数を1万回,ラッソ回帰をX_trainを説明変数,y_trainを目的変数として実行
print(las_reg_001.score(X_train,y_train)) #訓練データの精度を表示
print(las_reg_001.score(X_test,y_test)) #テストデータの精度を表示
#学習不足の典型的な例、どっちの精度も低い
```

```python
pd.DataFrame({"lin_reg":lin_reg.coef_,"rid_reg":rid_reg.coef_,"las_reg_001":las_reg_001.coef_})
#線形、リッジ、新しいラッソ回帰の係数を表示、少しラッソの係数が上がった
```
