```python
#モジュールpandasをpdという変数に収納
#モジュールnumpyをnpという変数に収納
#モジュールmatplotlib.pyplotをpltとしてインポート　
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100,1) #100行1列の乱数をXに格納
```

```python
X # 0から１までの一様分布
```

```python
 #　Xに３をかけ、５と乱数を加算しｙに収納
y =  3*X +5 +np.random.rand(100,1)
```

```python
plt.scatter(X,y)  #Xとｙの散布図を表示
#plt.scatter()は散布図を描く
```

```python
# sklearnのサブモジュールlinear_modelからLinearRegressionをインポート
from sklearn.linear_model import LinearRegression
```

```python
"""
直点を格納
LInerRegressionは説明変数の値から目的変数の値を予測することのできる、線形回帰分析の基本
fit()は線形回帰モデルの当てはめを行う


線形回帰とは、目的変数（予測したい値）を説明変数（目的変数を導くための変数）を用いて予測する手法
"""

lin_reg = LinearRegression().fit(X,y)
```

```python
#lin_reg.coef_は傾き, 
#lig_reg.intercept_は切片

lin_reg.coef_, lig_reg.intercept_
```

```python
X_new = np.array([[0],[1]]) #0と1の二重リストを作成
y_new = lin_reg.coef_*X_new+lin_reg.intercept_  #傾きにX_newをかけて切片を加算し、ｙに収納
plt.plot(X_new,y_new,"red")  #X_new,y_newの線を赤色で書く
plt.scatter(X,y)  #X_new,y_newの散布図を作成
```

```python
#datにuriage.csvを読み込み、0列目はインデックス、項目として扱ってうえで、収納 
dat = pd.read_csv("uriage .csv",index_col=0)
```

```python
dat #datを呼び出す
```

```python
x = dat.drop("uriage",axis=1) #全体からuriageの列だけ抜けたものをｘに収納
y = dat.uriage #uriageの列のみyに収納
```

```python
lin_reg = LinearRegression().fit(x,y) #yを目的変数し、ｘを説明変数とした線形回帰
```

```python
lin_reg.coef_, lin_reg.intercept_ #傾きと切片
```

```python
#X_forecastに0列目は項目として、uriage_forecast.csvを読み込み、収納する
X_forecast = pd.read_csv("uriage_forecast.csv",index_col=0)
```

```python
pred = lin_reg.predict(X_forecast) #説明変数X_forecastのみを線形回帰分析し、予想したものをpredに収納
```
